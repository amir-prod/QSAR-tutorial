# Suggested Student Report Template

This notebook accompanies the QSAR tutorial chapter: **Suggested Student Report Template**.

This chapter is mostly writing-focused. This notebook creates a simple Markdown report template
that students can copy into their GitHub repository or convert into a short report.

```python
from pathlib import Path

report_template = '''# QSAR Mini-Project Report

## Title
Write a short title describing the QSAR task.

## Introduction
What endpoint was predicted and why is it important?

## Dataset
Where did the data come from?
How many compounds were used?
What cleaning steps were performed?

## Methods
What descriptors or fingerprints were calculated?
What models were trained?
How was validation performed?

## Results
Include model performance tables and predicted vs. experimental plots.

## Discussion
Which model performed best?
Was there evidence of overfitting?
Which descriptors were important?
Are the results chemically meaningful?

## Limitations
Discuss dataset size, endpoint variability, applicability domain, and validation limitations.

## Conclusion
Summarize the main result in two or three sentences.
'''

output_path = Path("student_report_template.md")
output_path.write_text(report_template)

print("Saved:", output_path)
```
