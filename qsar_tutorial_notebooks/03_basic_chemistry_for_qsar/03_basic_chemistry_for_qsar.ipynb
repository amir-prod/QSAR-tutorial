## Visualizing molecules from SMILES with RDKit

```python
from rdkit import Chem
from rdkit.Chem import Draw

smiles = "CCO"  # ethanol
mol = Chem.MolFromSmiles(smiles)

Draw.MolToImage(mol)
```

## Visualize the tutorial SMILES examples

```python
from rdkit import Chem
from rdkit.Chem import Draw

molecule_data = [
    ("Methane", "C"),
    ("Ethanol", "CCO"),
    ("Acetic acid", "CC(=O)O"),
    ("Cyclohexane", "C1CCCCC1"),
    ("Benzene", "c1ccccc1"),
    ("Phenol", "c1ccccc1O"),
]

names = [name for name, smiles in molecule_data]
mols = [Chem.MolFromSmiles(smiles) for name, smiles in molecule_data]

Draw.MolsToGridImage(
    mols,
    molsPerRow=3,
    subImgSize=(280, 220),
    legends=names
)
```

## Load and visualize molecules from the shared dataset

```python
from pathlib import Path
import pandas as pd

def find_example_dataset():
    candidates = [
        Path("../data/example_qsar_dataset.csv"),
        Path("data/example_qsar_dataset.csv"),
        Path("../../data/example_qsar_dataset.csv"),
    ]
    for path in candidates:
        if path.exists():
            return path
    raise FileNotFoundError(
        "Could not find example_qsar_dataset.csv. "
        "Make sure the notebook is inside a chapter folder next to the shared data folder."
    )

data_path = find_example_dataset()
df = pd.read_csv(data_path)
df.head()
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

valid_rows = []
for _, row in df.iterrows():
    mol = Chem.MolFromSmiles(str(row["SMILES"]))
    if mol is not None:
        valid_rows.append((row["Name"], mol))

names = [name for name, mol in valid_rows[:12]]
mols = [mol for name, mol in valid_rows[:12]]

Draw.MolsToGridImage(
    mols,
    molsPerRow=4,
    subImgSize=(220, 180),
    legends=names
)
```
